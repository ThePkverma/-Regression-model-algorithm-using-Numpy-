```python
import numpy as np
import LinearRegerssion as lr

import matplotlib.pyplot as plt

n_samples = 1000
de_linearize = lambda X: np.cos(1.5 * np.pi * X) + np.cos( 5 * np.pi * X )
X = np.sort(np.random.rand(n_samples)) * 2
y = de_linearize(X) + np.random.randn(n_samples) * 0.1

clf=lr.LinearRegerssion(learning_rate=0.00000005,tolerance=0.5,lambda_=0.0,degree=4,maxiter=10**5)
error = clf.fit(X,y)
ypred = clf.predict(X)
plt.scatter(X,y)
plt.scatter(X,ypred)
```
